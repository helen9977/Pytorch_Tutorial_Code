```
import os
import glob
import random
import shutil
```

```
pwd = '/content/drive/My Drive/python/PyTorch_Tutorial'
dataset_dir = os.path.join(pwd, '..', '..', 'Data', 'cifar-10-png', 'raw_test')
train_dir = os.path.join(pwd, '..', '..', 'Data', 'train')
valid_dir = os.path.join(pwd, '..', '..', 'Data', 'valid')
test_dir = os.path.join(pwd, '..', '..','Data', 'test')

train_per = 0.8
valid_per = 0.1
test_per = 0.1
```

```
def makedir(new_dir):
  if not os.path.exists(new_dir):
    os.makedirs(new_dir)
```

```
for root, dirs, files in os.walk(dataset_dir): 
  # 返回的是一个三元组(root,dirs,files) 当前路径 / 文件夹 / 文件
  for sDir in dirs :
    imgs_list = glob.glob(os.path.join(root, sDir, '*.png')) # 满足条件的list
    random.seed(666) # 使 random() 生成的随机数是同一个。
    random.shuffle(imgs_list)
    imgs_num = len(imgs_list)

    train_point = int(imgs_num * train_per)
    valid_point = int(imgs_num * (train_per + valid_per))

    for i in range (imgs_num):
      if i < train_point:
        out_dir = os.path.join(train_dir, sDir)
      elif i < valid_point:
        out_dir = os.path.join(valid_dir, sDir)
      else:
        out_dir = os.path.join(test_dir, sDir)
      
      makedir(out_dir)
      # 以 "PATH" 中最后一个 '/' 作为分隔符，分隔后，将索引为0的视为目录（路径），将索引为1 / -1 的视为文件名
      out_path = os.path.join(out_dir, os.path.split(imgs_list[i])[-1])
      shutil.copy(imgs_list[i], out_path)

    print('Class:{},train:{},valid:{}, test:{}'.format(sDir, train_point,valid_point - train_point, imgs_num - valid_point))


```
